```python
cor_target = abs(corr["Class"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.35]
relevant_features
```

```python
x=pd.concat([data['feature_969'],data['feature_1048'],data['feature_1144'],data['feature_1154'],data['feature_1155'],data['feature_1199'],data['feature_1219'],data['feature_1244'],data['feature_1345'],data['feature_1400'],data['feature_1423'],data['feature_1425']],axis=1)
y=data['Class']
x.head()
```

```python
c=0
for i in data["feature_969"]:
    if i ==1:
        c=c+1
     
```

```python
c
```

```python
x.dtypes
```

```python
for column in x.columns:
    x[column] = (x[column] - x[column].min()) / (x[column].max() - x[column].min()) 
x.head()
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test ,y_train, y_test = train_test_split(x,y,
                                                   test_size=0.2,
                                                   random_state=0,
                                                  shuffle = True,
                                                  stratify = y)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_model=RandomForestClassifier()
forest_model.fit(X_train,y_train)
```

```python
random_pred=forest_model.predict(X_test)
```

```python
 y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
random = accuracy_score(y_test, random_pred)
```

```python
print ("Random Forest accuracy : ", round(random*100, 2))
```
